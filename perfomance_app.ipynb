```python
# Carregar o dataset de performance dos anúncios
path = 'raw/03-meta_ads_performance.json'
df = pd.read_json(file_path)

# Verificar as primeiras linhas para entender a estrutura
print(df.head(5))
```

```python
display(df.info())
```

```python
# Calcular a taxa de cliques (CTR): Impressões → Cliques
df['CTR'] = (df['clicks'] / df['impressions']) * 100

# Calcular a taxa de conversão (Cliques → Conversões)
df['conversion_rate'] = (df['conversions'] / df['clicks']) * 100

# Ver estatísticas de CTR e Taxa de Conversão
print(df[['CTR', 'conversion_rate']].describe())
```

```python
# Converter a coluna 'date' para datetime
df['date'] = pd.to_datetime(df['date'])

# Criar uma coluna de Mês/Ano
df['month_year'] = df['date'].dt.to_period('M')

# Agrupar os dados por mês/ano, somando as métricas
df_grouped = df.groupby('month_year').agg({
    'impressions': 'sum',
    'clicks': 'sum',
    'conversions': 'sum',
    'CTR': 'mean',
    'conversion_rate': 'mean'
}).reset_index()

print(df_grouped.head())
```

```python


# Gráfico de linha para visualizar o funil: Impressões, Cliques e Conversões
plt.figure(figsize=(10, 6))

# Plotando as métricas do funil
plt.plot(df_grouped['month_year'].astype(str), df_grouped['impressions'], label='Impressões', marker='o')
plt.plot(df_grouped['month_year'].astype(str), df_grouped['clicks'], label='Cliques', marker='o')
plt.plot(df_grouped['month_year'].astype(str), df_grouped['conversions'], label='Conversões', marker='o')

# Personalizando o gráfico
plt.title('Funil de Vendas por Mês/Ano', fontsize=16)
plt.xlabel('Mês/Ano', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
