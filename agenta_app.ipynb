```python
import pandas as pd
import plotly.express as ex
import matplotlib.pyplot as plt
```

```python
file1 = 'raw/01-agents.json'

if file1:
    df_agents = pd.read_json(file1)

```

```python
display(df_agents.head(5))
```

```python
for index, coluna in enumerate(df_agents.columns):
    print(index, coluna)
```

```python
# Agentes
display(df_agents['nome'])
```

```python

# Contar o número de agentes por região
agents_by_region = df_agents['region'].value_counts()

display(agents_by_region)
```

```python
# Converter a coluna 'hire_date' para datetime
df_agents['hire_date'] = pd.to_datetime(df_agents['hire_date'])

# Criar uma coluna para o mês/ano de contratação
df_agents['hire_month_year'] = df_agents['hire_date'].dt.to_period('M')

# Contar o número de contratações por mês/ano
hires_by_month = df_agents.groupby('hire_month_year').size()

print(hires_by_month)
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras para agentes por região
plt.figure(figsize=(10, 6))
agents_by_region.plot(kind='bar', color='lightblue')
plt.title('Distribuição de Agentes por Região', fontsize=16)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Número de Agentes', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de linha para tendências de contratação
plt.figure(figsize=(10, 6))
hires_by_month.plot(kind='line', marker='o', color='orange')
plt.title('Tendência de Contratação de Agentes por Mês/Ano', fontsize=16)
plt.xlabel('Mês/Ano', fontsize=12)
plt.ylabel('Número de Contratações', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
